````python
products=['t-shirt','mug','hat','book','keychain']

def initialize_inventory(products):
        up_inventory={product: int(input(f"Enter the quantity of {product}s available: "))for product in products}
        return up_inventory

def get_customer_order():
    number_of_prod=int(input('Enter the number of products you want to order:'))
    customer_order= {input('Enter the name of the product you want to order: ') for order in range(number_of_prod)}
    return customer_order

def calculate_order_statistics(customer_order, products):
    total = len(customer_order)
    percentage = (total / len(products)) * 100
    print("Order Statistics:")
    print(f"Total Products Ordered: {total}")
    print(f"Percentage of Unique Products Ordered: {percentage}")
    return total, percentage

def calculate_price(customer_order):
    price= sum(float(input(f'Enter the price of this product {product}:'))for product in customer_order)
    return price

def update_inventory(customer_order, inventory):
    updated_inventory={product:(value-1 if product in customer_order else value) for product, value in inventory.items() 
               if (value - 1 if product in customer_order else value) > 0}
    return updated_inventory

def print_updated_inventory(inventory):
    print("Updated inventory:")
    final_inventory=[ print(f"{product}: {quantity}") for product, quantity in inventory.items()]
    return final_inventory

inventory=initialize_inventory(products)
customer_order=get_customer_order()
calculate_order_statistics(customer_order, products)
inventory = update_inventory(customer_order, inventory)
print_updated_inventory(inventory)
total_price = calculate_price(customer_order)
print(f"Total Price: {total_price}")
````
